```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('DivisionAlgComparison.csv')
df.sort_values(by=['Numer of Individuals'])
X = df[df["Algorithm"] == "Moving Knife"]
Y = df[df.Algorithm == "Divide and Conquer"]


```

```python
ax = plt.gca()
plt.ylabel("Average Envy")
X.plot(kind='line',x ="Numer of Individuals", y = "Average Envy", label = "Moving Knife", ax=ax)
Y.plot(kind='line',x ="Numer of Individuals", y = "Average Envy", label = "Divide and Conquer", ax=ax)
plt.savefig('DivisionAlgComparison.png')
plt.show()
```
